Uma função chamada desenhaQuadrado que desenha um quadrado de lado 100


```python
def desenhaQuadrado(lado):
  #Inicie o display turtle
  t.initializeTurtle()
  for i in range(4):
    t.fd(lado)
    t.rt(90)
desenhaQuadrado(100)
```

Uma função chamada desenhaHexagono que desenha um hexagono de lado 100

```python
def desenhaHexagono(lado):
  #Inicie o display turtle
  t.initializeTurtle()
  for i in range(6):
  
      # Ande para frente por lado unidades
      t.forward(lado)
  
      # Vire 300 graus à esquerda
      t.left(300)
desenhaHexagono(100)
```

Uma função chamada desenhaCirculos, que desenha círculos que se intersectam de raio 100 (os centros dos círculos estão mais ou menos numa posição de triângulo equilátero), como exemplificado na figura abaixo.

```python
def desenhaCirculos(raio):
  #Inicie o display turtle
  t.initializeTurtle()
  t.circle(raio)
  t.up()
  t.goto(100, 0)
  t.down()
  t.circle(raio)
  t.up()
  t.goto(50, -100)
  t.down()
  t.circle(raio)

desenhaCirculos(100)
```

Forma alternativa, utilizando for e desenhando um triânguko subjacente, através da função foward

```python
def desenhaCirculos(raio):
  #Inicie o display turtle
  t.initializeTurtle()
  for i in range(3):
    
      # Vire 300 graus à esquerda
      t.down()
      t.circle(raio)
      t.up()
      
      # Ande para frente por 2 * raio unidades
      t.forward(raio)

      # Vire 30 graus à esquerda
      t.left(120)

desenhaCirculos(100)
```

Uma função chamada desenhaEmoticon, que desenha um “emoticon” sorrindo, como
abaixo:

```python
def desenhaEmoticon():
  t.initializeTurtle()
  def eye(col, rad):
      t.down()
      t.fillcolor(col)
      t.begin_fill()
      t.circle(rad)
      t.end_fill()
      t.up()
  
  
  # draw face
  t.circle(100)
  t.up()
  
  # draw eyes
  t.goto(-40, 120)
  eye('black', 15)
  t.goto(40, 120)
  eye('black', 15)
  
  # draw mouth
  t.goto(-40, 85)
  t.down()
  t.right(90)
  t.circle(40, 180)
  t.up()
desenhaEmoticon()
```
